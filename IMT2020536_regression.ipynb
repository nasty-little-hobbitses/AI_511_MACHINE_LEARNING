```python
class ORem: 
    def __init__(self, lq, uq):
        self.low_w = lq - 1.5*(uq - lq)
        self.high_w = uq + 1.5*(uq - lq)
    def removeOutlier(self, x):
        return (x if ((x <= self.high_w) & (x >= self.low_w)) 
        else (self.low_w if x < self.low_w else (self.high_w)))
```

```python
def OutRem(dt, columns):
    for i in columns:
        q1 = dt[i].quantile(0.25)
        q3 = dt[i].quantile(0.75)
        outlier_remover = ORem(q1, q3)
        dt[i] = dt[i].apply(outlier_remover.removeOutlier)
    return dt
```

```python
def train_test_split(data, split_ratio):
    train_data = pd.DataFrame()
    test_data = pd.DataFrame()
    data_size = data.shape[0]
    k = data.copy()
    k = k.sample(frac=1).reset_index(drop=True)
    test_data = k[0:int(split_ratio*data_size)].reset_index(drop=True)
    train_data = k[int(split_ratio*data_size):].reset_index(drop=True)
    x_train = train_data.iloc[:, 0:18]
    y_train = train_data["actual_productivity"]
    x_test = test_data.iloc[:, 0:18]
    y_test = test_data["actual_productivity"]
    return x_train, y_train, x_test, y_test
```
